```python
import pandas as pd
import numpy as np
import os
import sys
import json
```

```python
cats = ["Video_Games", "Pet_Supplies", "Home_and_Kitchen", "Health_and_Household", "Grocery_and_Gourmet_Food", "Tools_and_Home_Improvement"]

for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    valid_folder = f"data/raw/valid/"
    file_name = f"{cat}_valid.csv"

    # Read the CSV file
    df_valid = pd.read_csv(os.path.join(valid_folder, file_name))

    test_folder = f"data/raw/test/"
    file_name = f"{cat}_test.csv"
    # Read the CSV file

    df_test = pd.read_csv(os.path.join(test_folder, file_name))

    print(cat, "\n")
    print("Train shape:", df.shape)
    print("Valid shape:", df_valid.shape)
    print("Test shape:", df_test.shape)
    print()
```

```python
# Histogram of ratings for train data using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Train
for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    # Plot histogram
    sns.histplot(data=df, x='rating', bins=5, discrete=True, kde=False, edgecolor='black')
    plt.title(f"Rating distribution for {cat}")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    plt.show()
```

```python
# number of interactions per user in train data
# one table that has categories as rows and  min, max, mean, median, std as columns

# Create a DataFrame to store the results
results = pd.DataFrame(columns=["Category", "Min", "Max", "Mean", "Median", "Std"])
# Iterate through each category

for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    # Count interactions per user
    user_interactions = df.groupby('user_id').size()

    # Calculate statistics
    min_interactions = user_interactions.min()
    max_interactions = user_interactions.max()
    mean_interactions = user_interactions.mean()
    median_interactions = user_interactions.median()
    std_interactions = user_interactions.std()

    # vertically stack the results
    results = pd.concat([results, pd.DataFrame([[cat, min_interactions, max_interactions, mean_interactions, median_interactions, std_interactions]], columns=results.columns)], ignore_index=True)
```

```python
# Percentage of users per category in train data who have 1, 2, 3, 4, 5, 6-10, 11-20, 20+ interactions
# Create a DataFrame to store the results

results = pd.DataFrame(columns=["Category", "1", "2", "3", "4", "5", "6-10", "11-20", "20+"])

# Iterate through each category
for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    # Count interactions per user
    user_interactions = df.groupby('user_id').size()

    # Total number of users
    total_users = len(user_interactions)

    # Calculate percentages for each interaction range
    percent_1 = (user_interactions == 1).sum() / total_users * 100
    percent_2 = (user_interactions == 2).sum() / total_users * 100
    percent_3 = (user_interactions == 3).sum() / total_users * 100
    percent_4 = (user_interactions == 4).sum() / total_users * 100
    percent_5 = (user_interactions == 5).sum() / total_users * 100
    percent_6_10 = ((user_interactions >= 6) & (user_interactions <= 10)).sum() / total_users * 100
    percent_11_20 = ((user_interactions >= 11) & (user_interactions <= 20)).sum() / total_users * 100
    percent_20_plus = (user_interactions > 20).sum() / total_users * 100

    # Vertically stack the results
    results = pd.concat(
        [results, pd.DataFrame([[cat, percent_1, percent_2, percent_3, percent_4, percent_5, percent_6_10, percent_11_20, percent_20_plus]], 
        columns=results.columns)], 
        ignore_index=True
    )

# Set the category as the index
results = results.set_index("Category")

# Convert percentages to float with two decimal places
results = results.round(2)

# Display the results
results
```

```python
# Percentage of items per category in train data for which we have 1, 2, 3, 4, 5, 6-10, 11-20, 20+ interactions
# Create a DataFrame to store the results

results = pd.DataFrame(columns=["Category", "1", "2", "3", "4", "5", "6-10", "11-20", "20+"])
# Iterate through each category
for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    # Count interactions per item
    item_interactions = df.groupby('parent_asin').size()

    # Total number of items
    total_items = len(item_interactions)

    # Calculate percentages for each interaction range
    percent_1 = (item_interactions == 1).sum() / total_items * 100
    percent_2 = (item_interactions == 2).sum() / total_items * 100
    percent_3 = (item_interactions == 3).sum() / total_items * 100
    percent_4 = (item_interactions == 4).sum() / total_items * 100
    percent_5 = (item_interactions == 5).sum() / total_items * 100
    percent_6_10 = ((item_interactions >= 6) & (item_interactions <= 10)).sum() / total_items * 100
    percent_11_20 = ((item_interactions >= 11) & (item_interactions <= 20)).sum() / total_items * 100
    percent_20_plus = (item_interactions > 20).sum() / total_items * 100

    # Vertically stack the results
    results = pd.concat(
        [results, pd.DataFrame([[cat, percent_1, percent_2, percent_3, percent_4, percent_5, percent_6_10, percent_11_20, percent_20_plus]], 
        columns=results.columns)], 
        ignore_index=True
    )
```

```python
# Set the category as the index
results = results.set_index("Category")

# Convert percentages to float with two decimal places
results = results.round(2)

# Display the results
results
```

```python
# for each category number of unique users and items

# Create a DataFrame to store the results

results = pd.DataFrame(columns=["Category", "Unique Users", "Unique Items"])
# Iterate through each category

for cat in cats:
    train_folder = f"data/processed/train/"
    file_name = f"{cat}_train.csv"

    # Read the CSV file
    df = pd.read_csv(os.path.join(train_folder, file_name))

    # Count unique users and items
    unique_users = df['user_id'].nunique()
    unique_items = df['parent_asin'].nunique()

    # Vertically stack the results
    results = pd.concat(
        [results, pd.DataFrame([[cat, unique_users, unique_items]], columns=results.columns)], 
        ignore_index=True
    )
# Set the category as the index
results = results.set_index("Category")
# Convert percentages to float with two decimal places
results = results.round(2)
# Display the results
results
```
